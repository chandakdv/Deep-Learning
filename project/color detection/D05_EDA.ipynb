```python
from PIL import Image
import os
import glob

# Path to the folder containing images
folder_path = r"C:\Users\titik\OneDrive\Desktop\6th sem\DLL\Lab\Project\violet\*"

# Get a list of all JPEG files in the folder
jpg_files_before = glob.glob(os.path.join(folder_path, "*.jpg"))

# Count the number of JPG files before conversion
num_jpg_before = len(jpg_files_before)

# Iterate over each JPEG file
for jpg_file in jpg_files_before:
    # Open the image file
    with Image.open(jpg_file) as img:
        # Convert the image to RGB mode 
        img = img.convert("RGB")
        
        # Generate the output PNG filename by replacing ".jpg" with ".png"
        png_file = os.path.splitext(jpg_file)[0] + ".png"
        
        # Save the image as PNG, overwriting the original JPG file
        img.save(png_file, "PNG")
        
        # Remove the original JPG file
        os.remove(jpg_file)
        
        print(f"Converted {jpg_file} to {png_file} and replaced the original JPG file")

# Get a list of all PNG files in the folder after conversion
png_files_after = glob.glob(os.path.join(folder_path, "*.png"))

# Count the number of PNG files after conversion
num_png_after = len(png_files_after)

# Display the total number of JPG files before and after conversion
print(f"Total number of JPG files before conversion: {num_jpg_before}")
print(f"Total number of PNG files after conversion: {num_png_after}")

#after running it once jpg file will be shown as 0 only becaoz conversion ek baar hogya h already
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest'
)
```

```python
import cv2
import os

# Load the image
image_path = os.path.join(folder_path,"C:/Users/titik/OneDrive/Desktop/6th sem/DLL/Lab/Project/grey/gray3.jpg")  # Replace 'your_image.jpg' with your actual image filename
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print("Error: Unable to load image from", image_path)
else:
    # Your preprocessing and data augmentation code here
    # Example:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = image.reshape((1,) + image.shape)

    # Generate augmented images
    augmented_images = []
    for batch in datagen.flow(image, batch_size=1):
        augmented_images.append(batch[0].astype('uint8'))
        if len(augmented_images) >= 5:
            break

    plt.figure(figsize=(10, 5))
    for i in range(len(augmented_images)):
        plt.subplot(1, 5, i+1)
        plt.imshow(augmented_images[i])
        plt.axis('off')
plt.show()
```
